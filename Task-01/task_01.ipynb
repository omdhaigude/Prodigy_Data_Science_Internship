```python
import pandas as pd
df = pd.read_csv("population.csv", skiprows=4)
df = df[["Country Name", "2020"]]
df.head()
```

```python
import matplotlib.pyplot as plt
top10 = df.sort_values(by="2020", ascending=False).head(10)

plt.figure()
plt.bar(top10["Country Name"], top10["2020"])
plt.xlabel("Country")
plt.ylabel("Population in 2020")
plt.title("Top 10 Countries by Population (2020)")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df["2020"].dropna())
plt.xlabel("Population in 2020")
plt.ylabel("Number of Countries")
plt.title("Distribution of Country Populations (2020)")
plt.show()
```
